### import relvant libraries and the dataset

```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('sales.csv')
raw_data.head(10)
```

```python
inputs = raw_data.iloc[:,:-1].values
targets = raw_data.iloc[:,-1].values
```

```python
print(inputs.shape)
print(targets.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.1, random_state=0)
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(x_train, y_train)
```

```python
y_pred = model_1.predict(x_test)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, y_pred)
print('the accuracy of linear regression model is {:.2f}%'.format(accuracy*100))
```

### SVR Model

```python
from sklearn.svm import SVR
model_2 = SVR(kernel='rbf')
model_2.fit(x_train, y_train)
```

```python
y_pred = model_2.predict(x_test)
```

```python
accuracy = r2_score(y_test, y_pred)
print('the accuracy of SVR model is {:.2f}%'.format(accuracy*100))
```

### Decision tree regression model

```python
from sklearn.tree import DecisionTreeRegressor
model_3 = DecisionTreeRegressor(random_state=0)
model_3.fit(x_train,y_train)
```

```python
y_pred = model_3.predict(x_test)
```

```python
accuracy = r2_score(y_test, y_pred)
print('the accuracy of Decision tree regression model is {:.2f}%'.format(accuracy*100))
```

### Random forest regression model

```python
from sklearn.ensemble import RandomForestRegressor
model_4 = RandomForestRegressor(n_estimators=10,random_state=0)
model_4.fit(x_train,y_train)
```

```python
y_pred = model_4.predict(x_test)
```

```python
accuracy = r2_score(y_test, y_pred)
print('the accuracy of Random forest regression model is {:.2f}%'.format(accuracy*100))
```

#### from training the data by more than regression models  we observe that the random forest regression is the best model with accuracy 93.97%
